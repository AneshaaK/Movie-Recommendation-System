```python
# compute the frequent itemsets using fpgriowth from mlxtend
from mlxtend.frequent_patterns.fpgrowth import fpgrowth
frequent_itemsets = fpgrowth(df_fp_growth, min_support=0.3, use_colnames = True)

# print the frequent itemsets
frequent_itemsets
```

```python
len(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

# compute and print the association rules
association_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
association_rules
```

```python
association_rules.head()
```

```python
association_rules.tail()
```

```python
df_res = association_rules.sort_values(by=['lift'], ascending=False)
df_res.head()
```

```python
df_res["antecedents_len"] = df_res["antecedents"].apply(lambda x: len(x))
df_res.sort_values(by=['antecedents_len'], ascending=False)
```

```python
df_res["consequents_len"] = df_res["consequents"].apply(lambda x: len(x))
df_res.sort_values(by=['consequents_len'], ascending=False)
```

```python
df_MIB = df_res[df_res['antecedents'].apply(lambda x: next(iter(x)) == 'A Beautiful Mind')]
```

```python
df_MIB.head(10)
```

```python
df_res['antecedents'] = df_res['antecedents'].apply(list)
df_res['consequents'] = df_res['consequents'].apply(list)
```

```python
df_res=df_res.explode('consequents')
df_res.tail()
```
